```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import os
from io import BytesIO
from PIL import Image
import cv2
import re
```

```python
df = pd.read_csv("submission_2509_rn34.csv")
df2 = pd.read_csv("submission_2509_rn34_3.csv")
df3 = pd.read_csv("submission_2809_rn34_1.csv")
df4 = pd.read_csv("submission_0110_rn34_5.csv")
```

```python
df = df.merge(df2, on="UID").merge(df3, on="UID").merge(df4, on="UID", suffixes=["_z", "_k"])
df
```

```python
df["pred"] = np.sqrt(np.mean([df["growth_stage_x"] ** 2, df["growth_stage_y"] ** 2, 
                              df["growth_stage_z"] ** 2, df["growth_stage_k"] ** 2], axis = 0))
```
